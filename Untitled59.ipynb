```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
y = x**2
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-3,3,10)
y = np.sin(x)
plt.figure()
plt.plot(x, y, color='crimson', marker='o', linestyle='-.')
plt.show()
```

```python
x = np.linspace(0.1,100,20)
y = np.log10(x)
plt.figure()
plt.plot(x, y, color='sienna', marker='s', linestyle='dotted')
plt.show()
```

```python
x = np.linspace(0,100,20)
y = np.sqrt(x)
plt.figure()
plt.plot(x, y, color='green', marker='d', linestyle='-')
plt.grid()
plt.title('Raiz quadrada')
plt.xlabel('X')
plt.ylabel('$\sqrt{x}$')
plt.show()
```

```python
x = np.linspace(0,100,20)
y = np.sqrt(x)
z = x/10
plt.figure(figsize=(4,4))
plt.plot(x, y, color='green', marker='d', linestyle='-')
plt.plot(x, z, color='red', marker='x', linestyle='-.')
plt.grid()
plt.title('funes')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
